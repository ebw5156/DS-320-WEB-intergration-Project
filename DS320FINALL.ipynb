JACCARD SIMILARY

````
result.sample(10)#IMBD
````

````
df5.sample(5)#netflix
````

````
full_df = result.replace(r'^\s*$', np.nan, regex=True).dropna(axis=0)
````

````
full_external_df = df5.replace(r'^\s*$', np.nan, regex=True).dropna(axis=0)
````

````
#jaccard similarity
col_vals = set(full_df['IMBD_country'].unique())
for col in full_external_df.columns:
    ext_col_vals = set(full_external_df[col].unique())
    intersection_size = len(col_vals.intersection(ext_col_vals))
    union_size = len(col_vals.union(ext_col_vals))
    jaccard = intersection_size / union_size
    if jaccard != 0:
        print(col + ' has similarity score ' + str(jaccard))
````

FLEX MATCHER

````
!pip install flexmatcher
import flexmatcher
````

````
#mapping for flexmatcher
result_map = {'IMBD_name': 'title',
                 'IMBD_genre': 'listed_in	',
                 'IMBD_detail': 'description',
                 'IMBD_year': 'release_year',
                 'IMBD_time':'duration',
                 'IMBD_rating':'rating',
                 'IMBD_cast':'cast',
                 'IMBD_type':'type',
                 'IMBD_country':'country',
                 'IMBD_id':'show_id',
                 'IMBD_cast':'director',
                 #'IMBD_type':'date_added',
                 'IMBD_rate':'rating',
                 'IMBD_votes':'IMBD_votes'}
df5_map = {'title': 'IMBD_name',
                 'listed_in': 'IMBD_genre	',
                 'description': 'IMBD_detail',
                 'release_year':'IMBD_year',
                 'duration':'IMBD_time',
                 'rating':'IMBD_rating',
                 'cast':'IMBD_cast',
                 'type':'IMBD_type',
                 'country':'IMBD_country',
                 'show_id':'IMBD_id',
                 'director':'IMBD_cast',
                 'date_added':'IMBD_year',
                 'rating':'IMBD_rate',
                 'IMBD_votes':'IMBD_votes'}
````

````
#in order to get the datasets in thr right format for flexmatcher
df5=df5.astype(str)
result=result.astype(str)
df6=df6.astype(str)
````

````
#training on the netflix and IMBD datset schema 
schema_list = [df5, result]
mapping_list = [df5_map,result_map]
fm= flexmatcher.FlexMatcher(schema_list, mapping_list, sample_size=300)
fm.train()
````

````
#prediction on the disney datset
predicted_mapping = fm.make_prediction(df6)
predicted_mapping
````
